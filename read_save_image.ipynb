```python
import numpy as np
import cv2
```

```python
image = cv2.imread('anish1.jpg') # grayscale
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)

cv2.waitKey(1)
cv2.destroyAllWindows()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('anish1_gray_image.jpg', gray_image)
```
