```python
df['amount'] = df['amount']/1000
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X = df[['amount','age']] # input value X1, X2 (Independent Variable)
y = df['next_payment'] # binary classification (Dependent Variable)
```

```python
model = LogisticRegression(max_iter=500, random_state=42) # Try: random_state, solver, multi_class

model.fit(X, y)
```

```python
model.intercept_, model.coef_, model.score(X,y)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y[10:50:5], model.predict(X[10:50:5]))
```

```python
metrics.confusion_matrix(y[10:50:5], model.predict(X[10:50:5]))
```

```python
import numpy as np
```

```python
def calculate(index, amount, age):
    # prediction = 1 / (1 + e^(-(b0 + b1*x1 + b2*x2)))
    
    b = (model.intercept_[0] + model.coef_[0][0]*amount + model.coef_[0][1]*age)
    
    predict = 1/(1 + np.exp(-b))
    
    crisp = 0 if predict < 0.5 else 1
    
    yes = 'Yes' if crisp > 0.5 else ''
        
    return f"{index} - Amount: {amount*1000}, Age: {age}, Prob: {predict} - Crisp: {crisp} {yes}"
```

```python
import random
```

```python
# Calculate probability

for i in list(range(1,25)):
    print(calculate(i, amount=random.randrange(1, 6), age=random.randrange(27,42)))
```

```python
for index,row in df.iterrows():
    print(calculate(index, row.amount, row.age))
```
